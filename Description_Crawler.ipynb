```python
def get_gcontent():
    cand = []    
    dic_cand = []
    num = 0
    with open ("Sarcasm_News_Dataset_all_2.json",'a') as fa:
        with open("Sarcasm_Headlines_Dataset - test.json",'r') as f:
            content = f.readline()
            header={'content-type': 'text/html; charset=UTF-8','User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
            while content:
                text = json.loads(content)
                link = text['article_link']
                print(link)
        #         response=request.Request(link,headers=header)
#                 if "theonion" in link:
                num+=1
                html=grequests.get(link,headers=header,timeout=10,verify=False)
    #         result=html.read().decode('utf-8')
    #         print(html.content)
                cand.append(html)   
                dic_cand.append(text)
#                 else:
#                     others.append(link)
                if num==100:
                    gmap = grequests.map(cand)
                    for i in range(len(gmap)):
                        get_title2(gmap[i],dic_cand[i],fa)
                    cand = []
                    dic_cand = []
                    num=0
                    print("-----------------------")
                        
                content = f.readline()
                
            gmap = grequests.map(cand)
            for i in range(len(gmap)):
                get_title2(gmap[i],dic_cand[i],fa)
            cand = []
            dic_cand = []
            num=0
            print("-----------------------")


def get_title(rep,text,fa): 
    if rep:
        soup = BeautifulSoup(rep.text,"html.parser")
        temp = soup.find(class_ = "primary-cli cli cli-text")
        if temp:
            temp1 = temp.find("p")
            if temp1:                
                text['intro'] = p.clean(temp1.text)
                fa.write(json.dumps(text)+"\r")
                print(text['intro'])
            else:
                text['intro'] = ""
                fa.write(json.dumps(text)+"\r")
        else:
#             h_others.append(link)
            text['intro'] = ""
            fa.write(json.dumps(text)+"\r")

    else:
        text['intro'] = ""
        fa.write(json.dumps(text)+"\r")
        
def get_title2(rep,text,fa): 
    if rep:
        soup = BeautifulSoup(rep.text,"html.parser")
        temp = soup.find("meta",{"name":"twitter:description"})
        if temp:
            temp1 = temp.get('content')
            if temp1:                
                text['intro'] = p.clean(temp1)
                fa.write(json.dumps(text)+"\r")
                print(text['intro'])
            else:
                text['intro'] = ""
                fa.write(json.dumps(text)+"\r")
        else:
#             h_others.append(link)
            text['intro'] = ""
            fa.write(json.dumps(text)+"\r")

    else:
        text['intro'] = ""
        fa.write(json.dumps(text)+"\r")
        
#         h_others.append(link)  
```
