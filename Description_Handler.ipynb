```python
import json

intro_dict = {}

with open("Sarcasm_News_Dataset_all.json",'r') as f:
    content = f.readline()
    while content:
        text = json.loads(content)
        intro = text['intro']
        if intro in intro_dict:
            intro_dict[intro] += 1
        else:
            intro_dict[intro] = 1
        content = f.readline()
```

```python
sort_dict = dict(sorted(intro_dict.items(), key=lambda item: item[1],reverse = True))
```

```python
bad_des = ["The Onion brings you all of the latest news, stories, photos, videos and more from America's finest news source.",
           'The latest news on Congress, campaigns, elections, policy and everything politics from HuffPost.',
          'Read the latest headlines, news stories, and opinion from Politics, Entertainment, Life, Perspectives, and more.',
           'Dig into the real stories that impact our everyday lives across film, TV, music, books, arts and pop culture.',
           'Catch up on the latest news, photos, videos, and more on College.',
           'Sports news and opinion',
           'Catch up on the latest World News news, photos, videos, and opinion.']
```

```python
num = 0
with open("Sarcasm_News_Dataset_handled.json",'w') as fa:
    with open("Sarcasm_News_Dataset_all.json",'r') as f:
        content = f.readline()
        while content:
            text = json.loads(content)
            intro = text['intro']
            if intro in bad_des:
                num+=1
                text['intro'] = "" 
            fa.write(json.dumps(text)+"\r")
            content = f.readline()
```
